## Task : Facial Detection

tips: you can use any cascades below:

haarcascade_frontalface_alt.xml  

haarcascade_frontalface_alt2.xml 

haarcascade_frontalface_default.xml 

```python
fn = "itmo2.png"
I = cv2.imread(fn)
Igray = cv2.cvtColor(I,cv2.COLOR_BGR2RGB)

detector = cv2.CascadeClassifier()
# Be attention! Find the right path in your computer,
# I use conda to manage the package so the path might be different from yours
cascade_fn = cv2.samples.findFile("D:\Anaconda\envs\pytorch\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
detector.load(cascade_fn)

winScale = 1.07
neighbors = 10
faces = detector.detectMultiScale(Igray,scaleFactor = winScale,minNeighbors = neighbors)
Iout = Igray.copy ()
for (x , y , w , h ) in faces :
    Iout = cv2.rectangle ( Iout , (x , y , w , h ) ,(0 , 255 , 0) , 2)
print(str(faces.shape[0])+" faces have detected")
    
plt.figure(figsize = (18,18))
plt.imshow(Iout);plt.title("The Result of CascadeClassifier with windows scaleFactor = "+str(winScale)+" and windows size = "+str(neighbors))
plt.show()
```

## Task: Body Parts Detection

```python
# detect the faces
I = cv2.imread("jobs1.png")
Igray = cv2.cvtColor(I,cv2.COLOR_BGR2RGB)

detector = cv2.CascadeClassifier()
# Be attention! Find the right path in your computer,
# I use conda to manage the package so the path might be different from yours
cascade_fn = cv2.samples.findFile("D:\Anaconda\envs\pytorch\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
detector.load(cascade_fn)

winScale = 1.07
neighbors = 12
faces = detector.detectMultiScale(Igray,scaleFactor = winScale,minNeighbors = neighbors)
Iout = Igray.copy ()
for (x , y , w , h ) in faces :
    Iout = cv2.rectangle ( Iout , (x , y , w , h ) ,(0 , 255 , 0) , 2)
print(str(faces.shape[0])+" faces have detected")

# detect the eyes

# Load eyes cascade
eye_detector = cv2.CascadeClassifier()
cascade_fn = cv2.samples.findFile ("D:\Anaconda\envs\yolo\Lib\site-packages\cv2\data\haarcascade_eye_tree_eyeglasses.xml" )
eye_detector.load (cascade_fn)

# For each face use it as a ROI
# and detect eyes
for (x , y , w , h ) in faces :
    Iface_top = Igray [ y : y + h * 2 // 3 , x : x + w ]
    eyes = eye_detector . detectMultiScale (Iface_top , scaleFactor = 1.05)
    for ( x2 , y2 , w2 , h2 ) in eyes :
        Iout = cv2.rectangle ( Iout ,( x + x2 , y + y2 , w2 , h2 ) ,color = (255,0,0))
print(str(eyes.shape[0])+" eyes have detected")
        
# detect the mouth
# Load eyes cascade
mouth_detector = cv2.CascadeClassifier()
cascade_fn1 = cv2.samples.findFile ("D:\Anaconda\envs\pytorch\Lib\site-packages\cv2\data\haarcascade_eye.xml" )
mouth_detector.load (cascade_fn1)

for (x , y , w , h ) in faces :
    Imouth_top = Igray [ y + h * 2//3 : y + h , x : x + w ]
    mouths = mouth_detector . detectMultiScale (Imouth_top , scaleFactor = 1.05)
    for ( x2 , y2 , w2 , h2 ) in mouths :
        Iout = cv2.rectangle ( Iout ,( x + x2 , y + y2 + h * 2//3 , w2 , h2 ) ,color = (0,0,255))
if(mouths.size != 0):
    print(str(mouths.shape[0])+" mouth have detected")
else:
    print(str("0 mouth have detected"))

plt.figure(figsize = (8,8))
plt.imshow(Iout)
plt.show()
```

## Optional Task : Vedio Detection

```python
# read from the camera
import cv2

cap = cv2.VideoCapture(0)

## Open the vediostream
while(cap.isOpened()):
    ret, frame = cap.read()
#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    detector = cv2.CascadeClassifier()
    cascade_fn = cv2.samples.findFile("D:\Anaconda\envs\pytorch\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
    detector.load(cascade_fn)

    winScale = 1.08
    neighbors = 13
    faces = detector.detectMultiScale(frame,scaleFactor = winScale,minNeighbors = neighbors)
    for (x , y , w , h ) in faces :
        frame = cv2.rectangle ( frame , (x , y , w , h ) ,(0 , 255 , 0) , 2)
    
    
    cv2.imshow('frame',frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# Use the sample method to accerate the speed of processing
import cv2

cap = cv2.VideoCapture("beans.mp4")
idx = 0
freq = 4
while(True):
    idx += 1
    ret = cap.grab()
    if not ret:
         break
        
    if idx % freq == 1:
        ret, frame = cap.retrieve()
        if frame is None:    # exist broken frame
             break
         # do something else
        detector = cv2.CascadeClassifier()
        cascade_fn = cv2.samples.findFile("D:\Anaconda\envs\pytorch\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
        detector.load(cascade_fn)

        winScale = 1.13
        neighbors = 15
        faces = detector.detectMultiScale(frame,scaleFactor = winScale,minNeighbors = neighbors)
        for (x , y , w , h ) in faces :
            frame = cv2.rectangle ( frame , (x , y , w , h ) ,(0 , 255 , 0) , 2)
        cv2.imshow('frame',frame)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
